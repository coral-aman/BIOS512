```r
library('tidyverse')

college_sports <- readr::read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-03-29/sports.csv')
```

```r
college_sports %>% head()
```

```r
college_sports2 = college_sports |> filter(sports == 'Basketball' | sports =='Football')

p = ggplot(college_sports2 , aes(x = year, color = sports))
p = p + geom_bar(aes(weight = total_rev_menwomen)) + scale_y_continuous(labels = scales::dollar_format() , labs(y="Total Revenue Male and Female Sports")) + 
scale_x_continuous(labs(x = "Years")) + theme_light() + theme(axis.text.x = element_text(angle = 90))

p = p + facet_wrap(~ sports) 

p
```

### For my final project, I chose to analyze a college sports dataset from TidyTuesday. I imported the csv file and determined the variables I would focus on. I was specifically interested in total revenue from sports and the total count of participating athletes and how these numbers varied across years and colleges. For my first figure, I focusd on how the total revenue from both male and female sports has changed over time across all schools in the United States. I first created a new dataset filtered by basketball and football. I chose to display this data in bar charts faceted by sport (basketball and football). The chart above shows the total revenue collected from both male and female sports from 2015-2019 at colleges nationwide. I modified the theme of the charts to be light, adjusted the axis lables and angles, and added a geom color encoding to color the chart based on sport.
```r
college_sports3 = college_sports |> filter(institution_name == 'University of North Carolina at Chapel Hill')

p = ggplot(college_sports3 , aes(x = year))

p = p + geom_bar(aes(weight = partic_women)) + scale_y_continuous(labels = scales::comma_format() , labs(y="Total Female Athletes at UNC Chapel Hill")) + 
scale_x_continuous(labs(x = "Years")) + theme_light() + theme(axis.text.x = element_text(angle = 90))


p
```

```r
college_sports4 = college_sports |> filter( sports == 'Basketball' | sports =='Soccer') |> filter(institution_name == 'University of North Carolina at Chapel Hill')

p = ggplot(college_sports4 , aes(x = year, color = sports))

p = p + geom_bar(aes(weight = partic_men)) + scale_y_continuous(labels = scales::comma_format() , labs(y="Total Male Athletes at UNC Chapel Hill")) + 
scale_x_continuous(labs(x = "Years")) + theme_light() + theme(axis.text.x = element_text(angle = 90))

p = p + facet_wrap(~ sports) 

p
```
